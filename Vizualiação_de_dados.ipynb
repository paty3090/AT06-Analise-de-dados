```python
import pandas as pd
import seaborn as se
```

```python
# Importa os dados
tempo = pd.read_csv("/content/drive/MyDrive/tempo.csv")
```

```python
tempo.head()
```

```python
# Dá nomes às colunas
tempo.columns = ["Aparência", "Temperatura", "Umidade", "Vento","Jogo"]
```

```python
# Visualização os dados
tempo.head(10)
```

```python
# Visualiza a frequência de cada categoria na coluna 'Aperência'
pd.Series(tempo['Aparência']).value_counts()
```

```python
# Visualiza através de um gráfico de barra a frequência de cada categoria na coluna 'Aparência'
pd.Series(tempo['Aparência']).value_counts().plot.bar(color=['skyblue', 'lightcoral'])
```

Gráfico de dispersão

Um gráfico de dispersão é usado para visualizar a relação entre duas variáveis numéricas;
Cada ponto representa uma observação do dataset.

```python
# Visualisa com um gráfico de dispersão a relação entre Salário e Idade
se.scatterplot(x='Umidade', y='Temperatura', data=tempo)
```

scatterplot()

* Função do Seaborn para criar gráficos de dispersão;

* Ela é mais "inteligente" do que o plt.scatter() do matplotlib, pois permite automaticamente colorir os pontos por categorias, ajustar
estilos e legendas.

```python
# Visualiza a relação entre duas variáveis numéricas (Aparencia e Temperatura) e colore conforme a
#variável vento (indicando se o cliente Vento ou não)
'''
data=dataset : Define o conjunto de dados a ser usado no gráfico.
x='Aparência' : Define a variável numérica que será representada no eixo X.
y='Temperatura' : Define a variável numérica que será representada no eixo Y.
hue='Vento' : Define a variável categórica que será usada para colorir os pontos.
palette='Set1' : Define a paleta de cores a ser usada para colorir os pontos.
'Set1' é uma paleta de cores do Seaborn/Matplotlib composta por cores vibrantes
e bem distintas, ideal para diferenciar categorias.
Exemplo de cores em Set1: Vermelho, Azul, Verde, Roxo ... (até 9 cores).
'''
se.scatterplot(data=tempo, x='Aparência', y='Temperatura', hue='Vento', palette='Set1')
# Possível interpretação: "Clientes mais velhos e com
```

pairplot()

* Criar uma matriz de gráficos de dispersão (scatterplots), também conhecida como matriz de pares;

* O objetivo principal é visualizar todas as relações bivariadas possíveis entre um conjunto de variáveis numéricas e como essas relações se comportam conforme uma variável categórica;

* Na diagonal: gráficos de distribuição (histograma ou densidade) de cada variável;

* Fora da diagonal: gráficos de dispersão entre cada par de variáveis;

* Este comando é extremamente poderoso para uma visão rápida e completa das interações entre variáveis numéricas e o efeito de uma variável categórica sobre elas.

```python

se.pairplot(tempo[['Aparência', 'Temperatura', 'Umidade', 'Vento', 'Jogo']], hue='Vento', palette='husl')
```
